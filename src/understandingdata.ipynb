## Cleaning Data

# Check and Delete the outlier for Age

```python

df_cust['age'] = 2024 - df_cust['Year_Birth']
```

```python
num_people_over_100 = df_cust[df_cust['age'] > 100]
num_people_over_100
```

```python
# Remove the Rows that the age is more than 100 
df_cust = df_cust[df_cust['age'] <= 100]
```

# Check the Date of customer's enrollment with the company

```python
df_cust.loc[:, 'Dt_Customer'] = pd.to_datetime(df_cust['Dt_Customer'], format='%d-%m-%Y')
```

```python
df_cust["Dt_Customer"] = df_cust["Dt_Customer"].apply(
    pd.to_datetime, dayfirst=True, errors="coerce"
)
df_cust['YearofDt'] = df_cust['Dt_Customer'].dt.year
```

# Check the Null data of Income and Replace by median


```python
sns.histplot(df_cust.Income)
plt.title('Distrbution of income')
```

```python
# Replace Null data with Median
df_cust.loc[:, 'Income'] = df_cust['Income'].fillna(df_cust['Income'].median())
```

```python
# Identify to remove income outliers using IQR
Q1 = df_cust['Income'].quantile(0.25)
Q3 = df_cust['Income'].quantile(0.75)
IQR = Q3 - Q1
df_outlier_income = df_cust[(df_cust['Income'] < (Q1 - 1.5 * IQR)) | (df_cust['Income'] > (Q3 + 1.5 * IQR))]
df_outlier_income
```

```python
# Remove the outlier of income that is about 8 records 
# df_cust = df_cust[(df_cust['Income'] >= (Q1 - 1.5 * IQR)) & (df_cust['Income'] <= (Q3 + 1.5 * IQR))]
```

# Make data of Education and Marital Status Categorical

```python
# Convert categorical columns
df_cust.loc[:,'Education'] = df_cust['Education'].astype('category')
df_cust.loc[:,'Marital_Status'] = df_cust['Marital_Status'].astype('category')
```

# Add Some Columns / All Kids , TotalAcceptedCmp , Expenses

```python
df_cust.loc[:,'Kids'] = df_cust['Kidhome'] + df_cust['Teenhome']
df_cust.loc[:,'TotalAcceptedCmp'] = df_cust['AcceptedCmp1'] + df_cust['AcceptedCmp2'] + df_cust['AcceptedCmp3'] + df_cust['AcceptedCmp4'] + df_cust['AcceptedCmp5']
df_cust.loc[:,'Expenses'] = df_cust['MntWines'] + df_cust['MntFruits'] + df_cust['MntMeatProducts'] + df_cust['MntFishProducts'] + df_cust['MntSweetProducts'] + df_cust['MntGoldProds']
```
